```python
#How many times each shark invested the deal? 
df[['ashneer_deal','anupam_deal','aman_deal','namita_deal','vineeta_deal','peyush_deal','ghazal_deal']].sum()
```

```python
#Find the equity percent that each sharks gets?
ashneer=df[df['ashneer_deal']==1]['equity_per_shark'].sum()
anupam=df[df['anupam_deal']==1]['equity_per_shark'].sum()
aman=df[df['aman_deal']==1]['equity_per_shark'].sum()
namita=df[df['namita_deal']==1]['equity_per_shark'].sum()
vineeta=df[df['namita_deal']==1]['equity_per_shark'].sum()
peyush=df[df['peyush_deal']==1]['equity_per_shark'].sum()
ghazal=df[df['ghazal_deal']==1]['equity_per_shark'].sum()
equity_percent_es={'ashneer':ashneer,'anupam':anupam,'aman':aman,'namita':namita,'vineeta':vineeta,'peyush':peyush,'ghazal':ghazal}
dff=pd.DataFrame(data=equity_percent_es.items(),columns=["sharkname","equity"])
dff
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=dff,x="sharkname",y="equity")
plt.show()
```
