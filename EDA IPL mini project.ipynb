```python
#Plot a histogram of number of matches played in a season by players
plt.hist(data['Mat'])
plt.xlabel('Number of matches')
plt.ylabel('Number of player')
plt.show()
```

```python
#Plot the histogram of balls faced by players
plt.hist(data['BF'])
plt.xlabel('Number of BF')
plt.ylabel('Number of player')
plt.show()
```

```python
#Top 20 players with most runs in a season
r=data.sort_values("Runs",ascending=False)
r.head(20)
```

```python
#Top 30 players with highest number of sixes
sixes=data.sort_values("6s",ascending=False)
sixes.head(30)
```

```python
#Bar chart for top 20 players with best strike rate in a season.

bestsr=data.sort_values("SR",ascending=False)
bestsr1=bestsr.head(20)
plt.figure(figsize=(8,6))
plt.barh(bestsr1['Player'],bestsr1['SR'])
plt.ylabel('players')
plt.xlabel('strike rate')
plt.show()
```

```python
#Scatter plot of runs scored by a player v/s balls faced in a season.


plt.scatter(data['Runs'],data['BF'])
plt.xlabel('balls faced',fontsize='50')
plt.ylabel('runs',fontsize='50')
plt.show()
```

```python
#KDE plot for averages on a season.
sns.kdeplot(data['Avg'],x='Avg',color='Blue')
sns.set(font_scale=50)
```
